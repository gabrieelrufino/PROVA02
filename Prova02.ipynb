```python
# certo
def selection_sort(arr):
    for i in range(len(arr)-1):
        min_idx = i                    
        for j in range(i+1, len(arr)):   
            if arr[j] < arr[min_idx]:    
                min_idx = j              

        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return arr
```

# a)
##### min_idx é atualizado errado dentro de (min_idx = j), então sempre vira o ultimo indice.
##### falta if arr[j] < arr[min_idx]
##### sobrescreve o valor do arr[min_idx] antes de salvar, entao a troca nao acontcece.

```python
# b)
import random

# funcao 
def selection_sort(arr):
    for i in range(len(arr)-1):
        min_idx = i + 1
        for j in range(i, len(arr)):
            min_idx = j
        arr[min_idx] = arr[i]
        arr[i] = arr[min_idx]
    return arr

# lista
lista = [random.randint(0, 99) for _ in range(7)]
print("Lista gerada", lista)

# print
resultado = selection_sort(lista[:])
print("Resultado", resultado)
```

### o min_idx não faz comparações, entao todas as iterações percorrem ate o ultimo indice da lista, então o min_idx sempre termina como 6


```python
# d)

def find_max(lista):
    maior = lista[0]
    for elemento in lista[1:]:
        if elemento > maior:
            maior = elemento
    return maior

print("Maior elemento:", find_max(lista))
```
